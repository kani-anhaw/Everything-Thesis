```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_graphml('Synthetic graphs/data.graphml')
nodes=nx.number_of_nodes(G)
edges=nx.number_of_edges(G)
print("Nodes:", nodes)
print("Edges:", edges)

# Convert it to a directed graph
DG = G.to_directed()

# Find strongly connected components
scc = list(nx.strongly_connected_components(DG))

# Select the largest strongly connected component
largest_scc = max(scc, key=len)

# Create a subgraph with only the largest strongly connected component
X = DG.subgraph(largest_scc)


cc = nx.average_clustering(X)
spl = nx.average_shortest_path_length(X)
d = sum(dict(X.degree()).values()) / len(X)
print("Avg Clustering Coefficient:", cc)
print("Avg Degree:", d)
print("Avg Shortest Path length:", spl)


# degree distributuion
degree_sequence = [d for n, d in X.degree()]

# Count the occurrences of each degree
degree_counts = nx.degree_histogram(X)

# Plot the degree distribution
plt.bar(range(len(degree_counts)), degree_counts, width=0.80, color='b')

plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.savefig("LIN hist_default.png")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
def draw(G, measures, measurename, scale_factor=1000, node_scale=10):
    elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 5]
    esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 5]

    # Adjust the scale_factor and node_scale
    node_size = [v * scale_factor for v in measures.values()]

    pos = nx.kamada_kawai_layout(G, scale=scale_factor)

    nodes = nx.draw_networkx_nodes(G,
                                   pos, node_size=node_size,
                                   cmap=plt.cm.plasma,
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys(),
                                   linewidths=0.5,
                                   edgecolors='black',  # Add black edges around nodes for better visibility
                                   node_shape='o',  # Change node shape to circle
                                   )

    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))

    nx.draw_networkx_edges(G, pos, edgelist=elarge, width=0.5)
    nx.draw_networkx_edges(G, pos, edgelist=esmall,
                           width=0.5, alpha=0.5,
                           edge_color='blue',
                           style='dashed')

    plt.title(measurename)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
    plt.savefig('sample_graph.png', dpi=300)

# Example usage
# draw(G, measures, "Your Measure Name", scale_factor=1000, node_scale=10)


draw(G, nx.degree_centrality(G),  'Degree Centrality')
```

```python
# BA algo functions

def rand_prob_node():
    nodes_probs = []
    for node in G.nodes():
        node_degr = G.degree(node)
        #print(node_degr)
        node_proba = node_degr / (2 * len(G.edges()))
        #print("Node proba is: {}".format(node_proba))
        nodes_probs.append(node_proba)
        #print("Nodes probablities: {}".format(nodes_probs))
    random_proba_node = np.random.choice(G.nodes(),p=nodes_probs)
    #print("Randomly selected node is: {}".format(random_proba_node))
    return random_proba_node

def add_edge():
        if len(G.edges()) == 0:
            random_proba_node = 0
        else:
            random_proba_node = rand_prob_node()
        new_edge = (random_proba_node, new_node)
        if new_edge in G.edges():
            print("!   ")
            add_edge()
        else:
            print("! ")
            G.add_edge(new_node, random_proba_node)
            print("Edge added: {} {}".format(new_node + 1, random_proba_node))
```
