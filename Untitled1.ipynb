```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to calculate clustering coefficient and plot degree distribution
def calculate_and_plot(graph, label, color):
    cc = nx.average_clustering(graph)
    print(f"Avg Clustering Coefficient for {label}:", cc)

    # Plot the degree distribution
    degree_sequence = [d for n, d in graph.degree()]
    degree_counts = nx.degree_histogram(graph)
    plt.plot(range(len(degree_counts)), degree_counts, label=f"{label} - CC: {cc:.4f}", color=color)

# Load the first graph
G = nx.read_graphml('Synthetic graphs/data.graphml')
DG = G.to_directed()
largest_scc = max(nx.strongly_connected_components(DG), key=len)
X = DG.subgraph(largest_scc)

# Load the second graph
sw = nx.watts_strogatz_graph(len(G), 4, 0.1)
largest_cc_sw = max(nx.connected_components(sw), key=len)
largest_cc_graph_sw = sw.subgraph(largest_cc_sw)

# Load the third graph
sf = nx.barabasi_albert_graph(len(G), 4)

# Load the fourth graph
rn = nx.erdos_renyi_graph(len(G), 0.001086)

# Plot the clustering coefficient distribution using line graph
calculate_and_plot(X, "Learning Interaction Network", 'b')
calculate_and_plot(largest_cc_graph_sw, "Small-world Network", 'g')
calculate_and_plot(sf, "Scale-free Network", 'r')
calculate_and_plot(rn, "Random Network", 'purple')

# Customize the plot
# plt.title('Degree Distribution Comparison')
plt.xlabel('Node')
plt.ylabel('Clustering Coefficient')
plt.xlim(0, 20)
plt.legend()
plt.show()
```
